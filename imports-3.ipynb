```python
import networkx as nx
import matplotlib.pyplot as plt
df2 = pd.read_csv('/Users/christhomas/Port Data/imports.csv', usecols = ['FR_ORIG','DMS_ORIG','DMS_DEST'])
G = nx.from_pandas_edgelist(df2,source = 'FR_ORIG',target = 'DMS_ORIG', edge_attr=None, create_using=nx.DiGraph())
G
H = nx.from_pandas_edgelist(df2,source = 'DMS_ORIG',target = 'DMS_DEST', edge_attr=None, create_using=nx.DiGraph())
H
C = nx.compose(G,H)
```

```python
nx.draw(G)
```

```python
nx.draw(C)
plt.show()
```
