```python
import pandas as pd
df1 = pd.read_csv('/Users/christhomas/Port Data/imports.csv')
df1
#imports the foreign port origin, domestic port origin, and domestic port destination FAF commodity import data
```

```python
df1.drop('FR_INMODE', axis = 1, inplace = True)
```

```python
df1.columns = ['node.0','node.1','node.2','commodity','dom_mode','distance','k_ton','value_mil']
```

```python
df1
```

```python
adj_df1 = df1.drop(df1[df1['value_mil'] == 0].sample(frac=1).index)
```

```python
adj_df1
```

```python
nodes = [col for col in adj_df1 if col.startswith('node.')]
node_df = pd.melt(adj_df1, id_vars = ['commodity','dom_mode','distance','k_ton','value_mil'], value_vars = nodes, var_name = 'Node', value_name = 'Location')
node_df
```
